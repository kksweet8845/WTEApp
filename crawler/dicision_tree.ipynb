# get all categories

```python
import numpy as np
import pandas as pd
```

```python
doc_ref = db.collection('love2eat')
docs = doc_ref.stream()
```

```python
categories = []

for doc in docs:
    tmp = doc.to_dict()
    categories.extend(tmp['categories'])
    print(tmp['categories'])
categories = np.unique(categories)
```

```python
categories
```

```python
df = pd.DataFrame(columns=categories)
```

```python
df
```

```python
doc_ref = db.collection('love2eat')
docs = doc_ref.stream()
for doc in docs:
    tmp = doc.to_dict()
    t = [False]*len(categories)
    for i, category in enumerate(categories):
        if category in tmp['categories']:
             t[i] = True
    tmp_df = pd.DataFrame([t], columns=categories)
    df = df.append(tmp_df, ignore_index=True)
```

# Generate decision tree

```python
from sklearn import tree
from sklearn.tree import export_graphviz
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(df.iloc[:, :], [ x for x in range(len(df))])
```

```python
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=df.columns[:],
                            filled=False,
                           proportion=True,
                           rotate=True,
                          rounded=True)
```

```python
import graphviz
graph = graphviz.Source(dot_data)
graph
```
